```python
import numpy as np
import matplotlib.pyplot as plt

# Suppose we want to obtain the fitted line to describe the relationship between x=[1,2,3,4,5] and y=[3,8,5,14,19]

def gradient_descent_matrix_method(X, y, learning_rate=0.01, iterations=10000):
    # Number of data points (m) and number of features (n)
    m, n = X.shape
    
    # Initialize parameters (theta) to zeros (one for each feature, including intercept)
    theta = np.zeros(n)
    
    for i in range(iterations):
        # Compute predictions: X.dot(theta) is a matrix-vector multiplication
        y_predicted = X.dot(theta)
        
        # Compute the gradient: (1/m) * X.T.dot(y_predicted - y)
        gradient = (1/m) * X.T.dot(y_predicted - y)
        
        # Update parameters (theta)
        theta = theta - learning_rate * gradient
    
    # Return the final parameters (theta)
    return theta

# Example usage
# Features (X): Add a column of ones for the intercept
# First column is 1 (intercept term), second column is the feature x
X = np.array([
    [1, 1],  # 1st data point: 1 (intercept term), x = 1
    [1, 2],  # 2nd data point: 1 (intercept term), x = 2
    [1, 3],  # 3rd data point: 1 (intercept term), x = 3
    [1, 4],  # 4th data point: 1 (intercept term), x = 4
    [1, 5]   # 5th data point: 1 (intercept term), x = 5
])

# Target values (y)
y = np.array([3, 8, 5, 14, 19])

# Run gradient descent with the matrix method
theta = gradient_descent_matrix_method(X, y)

# Final intercept and slope
intercept = theta[0]  # θ_0
slope = theta[1]      # θ_1

print(f"Intercept (θ_0) = {intercept}, Slope (θ_1) = {slope}")

# Plotting the data points (scatter plot)
plt.scatter(X[:, 1], y, color="blue", label="Data Points")

# Plotting the best-fitted line
x_values = np.linspace(1, 5, 100)  # Generate 100 points between 1 and 5 for a smooth line
y_values = intercept + slope * x_values  # Use the slope and intercept from the model
plt.plot(x_values, y_values, color="red", label="Fitted Line")

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Best Fitted Line')
plt.legend()

# Display the plot
plt.show()
```
